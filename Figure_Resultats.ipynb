# Bateau pointe triangulaire


```python
data_triangle1=np.load("Resultats/Matrices_Champ/SER/SERV2/SER1001_triangle_pi4.npy")
data_triangle2=np.load("Resultats/Matrices_Champ/SER/SERV2/SER1001_triangle_pi25.npy")

data_triangle3=np.load("Resultats/Matrices_Champ/SER/SERV2/SER_triangle_pi6.npy")

Dcm=23
L=5
coeff1=(Dcm-L*(1+1/2/np.tan(np.pi/8)))/(Dcm-(L/(12*np.tan(np.pi/8)*(4*np.tan(np.pi/8)+1))))
coeff1
coeff2=(Dcm-L*(1+1/2/np.tan(np.pi/5)))/(Dcm-(L/(12*np.tan(np.pi/5)*(4*np.tan(np.pi/5)+1))))
coeff2

coeff3=(Dcm-L*(1+1/2/np.tan(np.pi/12)))/(Dcm-(L/(12*np.tan(np.pi/12)*(4*np.tan(np.pi/12)+1))))


data_triangle1=data_triangle1*coeff1**3
data_triangle2=data_triangle2*coeff2**3
data_triangle3=data_triangle3*coeff3**3
```

```python

plt.figure(figsize=(9,3))
plt.plot(omega/2/np.pi,data_triangle1, lw=3,label="Pointe triangulaire:$45\degree$", color="blue")
plt.plot(omega/2/np.pi,data_triangle2,lw=3, ls="-", label="Pointe triangulaire $75\degree$", color="red")
plt.plot(omega/2/np.pi,data_triangle3,lw=3, ls="-", label="Pointe triangulaire $30\degree$", color="orange")
plt.xlabel("Fréquence(Hz)",fontsize=12, fontweight="bold")
plt.ylabel("SER",fontsize=12, fontweight="bold")
plt.yticks(fontsize=12)
plt.xticks(fontsize=12)
#plt.ylim(0,3.0)
plt.hlines(np.mean(data_triangle1),0,10000,label="SER moyenne: $45\degree$",linestyle="--", color="blue")
plt.hlines(np.mean(data_triangle2),0,10000,label="SER moyenne: $75\degree$",linestyle="--", color="red")
plt.hlines(np.mean(data_triangle3),0,10000,label="SER moyenne: $30\degree$",linestyle="--", color="orange")
plt.legend(ncol=2,fontsize=11)
plt.savefig("Resultats/SER_Triangle.png",dpi=200,bbox_inches="tight")
plt.show()

np.mean(data_triangle1),np.mean(data_triangle3)

```

# Sweep triangle 

## Source négative 

```python
def Correction(alpha):
    Dcm=22.25
    L=5
    return (Dcm-L*(1+1/2/np.tan(alpha/2)))/(Dcm-(L/(12*np.tan(alpha/2)*(4*np.tan(alpha/2)+1))))

```

```python
coeff_arr=np.linspace(np.pi/8, np.pi/1.1,16)
omega_arr=2*np.pi*np.linspace(100,10000,50)



Results=np.load("Resultats/Matrices_Champ/SER/SERV2/Sweep_Trianglepi8api1p1v2.npy")
C2=np.transpose(Correction(coeff_arr)**2)
Results=np.transpose((C2*np.ones([50,16])))*Results
plt.figure(figsize=(7,7))

plt.imshow(np.log10(np.array(Results)), cmap="jet",interpolation="none",aspect="auto")

plt.xticks(np.arange(0,omega_arr.shape[0],4),np.round(omega_arr[::4]/2/np.pi,),rotation=20)
plt.xlabel("Fréquence (Hz) ",fontsize=14)

plt.yticks(np.arange(coeff_arr.shape[0]),np.round(np.rad2deg(coeff_arr)),fontsize=12)
plt.ylabel("Angle(deg)", fontsize=14)

plt.title("Log de la SER en fonction du angle \n du triangle et de la fréquence ", fontsize=12)

plt.show()


```
