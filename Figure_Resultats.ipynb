```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import patches
```

```python
Nx=1501
Lx = 40
# Espace entre chaque noeud
dx = (Lx) / (Nx - 1)

# Nombre de points en y
Ny = Nx
# Longueur en x (m)
Ly = Lx
# Espace entre chaque noeud
dy = dx
h=dx
# Épaisseur (en points de la couche de PML)
N_PML = 10

centre_bois_x = int(Nx / 2)  # 65
centre_bois_y = int(1.3 * Nx / 2)  # 600#60
# Longueur en x,y du bois (en points)
Nx_Bois = int(5 / dx)
Ny_Bois = int(10 / dx)

# Emplacement de la source
S_x = int(Nx/2)
S_y = 75


```

# Pour les carrés

```python
data=np.transpose(np.load("Resultats/Matrices_Champ/Carre1501_f250Hz_SER0.291405226956879.npy"))[N_PML:-N_PML,N_PML:-N_PML]
plt.imshow(np.real(data), cmap="jet")
NxM=Nx-2*N_PML
plt.fill_between([(NxM-Nx_Bois)/2,(NxM+Nx_Bois)/2],(centre_bois_y-2*N_PML)-(Ny_Bois)/2,(centre_bois_y-2*N_PML)+(Ny_Bois)/2, edgecolor="black",\
                lw=3,facecolor="gray",hatch="o")
plt.xticks(np.arange(0,NxM,200),np.arange(0,NxM*dx,5,dtype=np.int), fontsize=12)
plt.yticks(np.arange(0,NxM,200),np.arange(0,NxM*dx,5,dtype=np.int), fontsize=12)
plt.xlabel("x (en m.)", fontsize=12, fontweight="bold")
plt.ylabel("y (en m.)", fontsize=12, fontweight="bold")
patches.Circle((10,10),radius=100)
circle=plt.Circle((S_x-N_PML,S_y-N_PML),1/dx,color="r",fill=False)
ax=plt.gca()
ax.add_artist(circle)
plt.savefig("Resultats/Carre_250Hz.png",dpi=200,bbox_inches="tight")
# ax2 = plt.axes([0.7, 0.55, 0.155, 0.32],)
# ax2.

```

# Bateau rectangle 

```python
data_carre=np.load("Resultats/Matrices_Champ/SER/SERV2/SER1001_Carre.npy")
data_rectangle=np.load("Resultats/Matrices_Champ/SER/SERV2/SER1001_Rectangle.npy")
omega=np.load("Resultats/Matrices_Champ/SER/SERV2/Omega1001_cercle_10.npy")
```

```python

plt.figure(figsize=(6,4))
plt.plot(omega/2/np.pi,data_carre, lw=3,label="Bateau carré: 10 m x 10 m", color="blue")
plt.plot(omega/2/np.pi,data_rectangle,lw=3, ls="-", label="Bateau rectangulaire: 5 m x 10 m", color="orange")
plt.xlabel("Fréquence(Hz)",fontsize=12, fontweight="bold")
plt.ylabel("SER",fontsize=12, fontweight="bold")
plt.yticks(fontsize=12)
plt.xticks(fontsize=12)

plt.hlines(np.mean(data_carre),0,10000,label="SER moyen: Bateau carré",linestyle="--", color="blue")
plt.hlines(np.mean(data_rectangle),0,10000,label="SER moyen: Bateau rectangulaire",linestyle="--", color="orange")
plt.legend(fontsize=11)
plt.savefig("Resultats/SER_Rectangle.png",dpi=200,bbox_inches="tight")
plt.show()
```

# Sweep

```python
coeff_arr=np.linspace(1.0, 4.0,16)
omega_arr=2*np.pi*np.linspace(100,10000,50)


Results=np.load("Resultats/Matrices_Champ/SER/SERV2/Sweep_Cercle1a4.npy")
plt.figure(figsize=(8,8))

plt.imshow(np.log10(np.array(Results)), cmap="jet",interpolation="hanning",aspect="auto")

plt.xticks(np.arange(0,omega_arr.shape[0],4),np.round(omega_arr[::4]/2/np.pi,),rotation=20)
plt.xlabel("Fréquence (Hz) ",fontsize=14)

plt.yticks(np.arange(coeff_arr.shape[0]),np.round(coeff_arr,1),fontsize=12)
plt.ylabel("Rayon= Coeff*Largeur/2", fontsize=14)

plt.title("Log de la SER en fonction du coefficient \n du cercle et de la fréquence ", fontsize=12)

plt.show()
```

# Sweep triangle 



```python
coeff_arr=np.linspace(np.pi/8, np.pi/1.1,16)
omega_arr=2*np.pi*np.linspace(100,10000,50)


Results=np.load("Resultats/Matrices_Champ/SER/SERV2/Sweep_Trianglepi8api1p1v2.npy")
plt.figure(figsize=(7,7))

plt.imshow(np.log10(np.array(Results)), cmap="jet",interpolation="none",aspect="auto")

plt.xticks(np.arange(0,omega_arr.shape[0],4),np.round(omega_arr[::4]/2/np.pi,),rotation=20)
plt.xlabel("Fréquence (Hz) ",fontsize=14)

plt.yticks(np.arange(coeff_arr.shape[0]),np.round(np.rad2deg(coeff_arr)),fontsize=12)
plt.ylabel("Angle(deg)", fontsize=14)

plt.title("Log de la SER en fonction du angle \n du triangle et de la fréquence ", fontsize=12)

plt.show()
```
